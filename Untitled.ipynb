### Data Representation TTA

#### Task

- Clean raw data to get hourly Google Trends data on bitcoin

#### Time

- Took me about ~ 1 h 30 mins in total

#### Idea
- As descirbed in the problem, value_hour is rescaled with respect to week and value_week is rescaled with respect to month, so we can use value_week and value_month as scalling factors to get real value_hour
- merge weekly data and monthly data onto hourly data and foward fill NA values to propagate the scalling factors. Then multiply all three values together.

```python
import pandas as pd
# pd.set_option('display.max_rows', None)
```

```python
df_hour = pd.read_csv('hourly_data.csv')
df_hour['date'] = pd.to_datetime(df_hour['date'])
df_hour
```

```python
df_week = pd.read_csv('weekly_data.csv')
df_week['date'] = pd.to_datetime(df_week['date'])
df_week
```

```python
df_month = pd.read_csv('monthly_data.csv')
df_month['date'] = pd.to_datetime(df_month['date'])
df_month
```

```python
sum(df_hour['value_hour']==100)
```

```python
df_month.info()
```

```python
df_hour=df_hour.merge(df_week,how='left', on='date')
df_hour=df_hour.merge(df_month,how='left', on='date')
df_hour
```

```python
df_hour = df_hour[['value_hour', 'date','value_week','value_month']]
df_hour
```

```python
df_hour=df_hour.fillna(method='ffill')
df_hour
```

```python
df_hour['value_hour'] = df_hour['value_hour']*df_hour['value_week']*df_hour['value_month']/10000
df_hour
```

```python
import matplotlib.pyplot as plt
plt.plot(df_hour['date'], df_hour['value_hour'],'-')
```

```python
df_hour['date'] = df_hour['date'].dt.strftime('%Y-%m')
df = df_hour.groupby('date').mean()
df
```

```python
plt.plot(df.index, df['value_hour'])
```
